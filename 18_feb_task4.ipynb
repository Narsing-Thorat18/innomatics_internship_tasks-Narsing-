**Problem 1: Mobile Recharge Validation**

```python
def validate_recharge(amount):
    plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount should be 50 or more")
        return False

    if amount in plans:
        print("Recharge successful")
        return True
    else:
        print("Invalid plan")
        print("Valid plans are 199, 299, 399, 599")
        return False


# using while loop for retry
while True:
    try:
        amt = int(input("Enter recharge amount: "))

        result = validate_recharge(amt)

        if result == True:
            break
        else:
            print("Try again")

    except:
        print("Enter numbers only")
```

**Problem 2: Inventory Reorder Alert System**

```python
def check_stock(products):

    for item in products:
        if products[item] < 15:
            print(item, "- Reorder Alert")
        else:
            print(item, "- Stock OK")
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Soap": 25
}

check_stock(inventory)
```

**Problem 3: Student Result Processing System**

```python
def check_result(marks):

    total = 0
    for m in marks:
        total = total + m

    average = total / len(marks)

    print("Average marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


#marks list
student_marks = [60, 45, 70, 50, 40]

result = check_result(student_marks)

print("Result:", result)
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):

    base_fare = 50
    per_km = 12

    total = base_fare + (distance * per_km)

    if peak == "yes":
        extra = total * 0.25
        total = total + extra

    return total
# retry using while loop
while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_time = input("Is it peak hour? (yes/no): ")

        fare = calculate_fare(km, peak_time)

        print("Total Fare is:", fare)

        break

    except:
        print("Invalid input. Try again.")
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
def check_attendance(att_list):

    total_days = len(att_list)
    present_days = 0

    #count number of "P"
    for day in att_list:
        if day == "P":
            present_days = present_days + 1

    percentage = (present_days / total_days) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


#attendance list
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_attendance(attendance)
print("Result:", result)
```

**Problem 6: Password Strength Checker**

```python
def check_password(password):
    if len(password) < 8:
        return "Weak Password"

    digit_found = False
    special_found = False
    for i in password:
        if i.isdigit():
            digit_found = True

        if i == "@" or i == "#" or i == "$":
            special_found = True

    # check conditions
    if digit_found and special_found:
        return "Strong Password"
    else:
        return "Weak Password"


# take input from user
pwd = input("Enter your password: ")

result = check_password(pwd)

print("Result:", result)
```
