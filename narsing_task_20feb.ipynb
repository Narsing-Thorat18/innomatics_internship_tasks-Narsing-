Problem 1: Social Media – Post Engagement Analyzer


```python
def analyze_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"


    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes_data = [200, 150, 300, 250, 150]
analyze_engagement(likes_data)
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_checker(rainfall_data, required_level):
    total = 0

    #total rainfall
    for rain in rainfall_data:
        total += rain

    #average rainfall
    average = total / len(rainfall_data)

    # Check rainfall adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

seasonal_rainfall = [70, 75, 80, 65, 70]
required = 70

rainfall_checker(seasonal_rainfall, required)
```

Problem 4: Social Media – Duplicate Account Detection


```python
def detect_duplicates(usernames):
    unique_users = set()
    duplicate_found = False

    for user in usernames:
        if user in unique_users:
            duplicate_found = True
            break
        unique_users.add(user)

    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example
user_list = ["alice", "bob", "charlie", "alice"]
detect_duplicates(user_list)
```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(prices, threshold=2000):
    premium_crops = []

    for price in prices:
        if price > threshold:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)
crop_prices = [1500, 2500, 1800, 3200, 1900]
filter_premium_crops(crop_prices)
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Loop to count present days
    for day in attendance_list:
        if day == "Present":
            present_days += 1

    #attendance percentage
    percentage = (present_days / total_days) * 100

    #eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_data = [
    "Present", "Present", "Absent", "Present", "Present",
    "Present", "Present", "Absent", "Present", "Present"
]

check_exam_eligibility(attendance_data)
```

Problem 10: Smart Electricity Bill Analyzer


```python
def calculate_electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", status)
```
