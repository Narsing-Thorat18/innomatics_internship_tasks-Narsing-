```python
#question 1
#user login check
correct_username = "admin"
correct_password = "1234"
#input
username = input("Enter username: ")
password = input("Enter password: ")
# check login
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#question 2
#student mark list
marks = [45, 78, 90, 33, 60]

#counters
pass_count = 0
fail_count = 0

# checking each student's marks
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

# printing result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
#question 3
# original list of names
names = [" Alice ", "bob", " CHARLIE "]

# empty list to store cleaned names
cleaned_names = []

# cleaning
for name in names:
    cleaned = name.strip().lower()
    cleaned_names.append(cleaned)

#result
print("Cleaned Names List:", cleaned_names)
```

```python
#question 4
# list of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# check each message
for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)

    if length > 10:
        print("Status: Long message")
    else:
        print("Status: Short message")

    print()
```

```python
#question 5
# list of log messages
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# counter for errors
error_count = 0

# check each log entry
for log in logs:
    if log == "ERROR":
        error_count += 1

# print result
print("Total ERROR messages:", error_count)
```
