Problem Statement 1:Smart Parking Lot Management System

```python
# Smart Parking Lot Management System
def smart_parking_system(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0

    # Loop through logs
    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1

        # Track peak parking usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    # Check parking status
    if current_parked > capacity:
        status = "ALERT: Parking Exceeded Capacity!"
    elif current_parked == capacity:
        status = "Parking Full"
    else:
        status = "Available"

    # Display results
    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Run the function
smart_parking_system(capacity, vehicle_logs)
```

**Problem Statement 2:Online Food Delivery Time Estimator**

```python
def estimate_delivery_time(distance, traffic, weather):

    base_speed = 30
    base_time = (distance / base_speed) * 60
    traffic = traffic.lower()
    if traffic == "high":
        base_time *= 1.6
    elif traffic == "medium":
        base_time *= 1.3
    elif traffic == "low":
        base_time *= 1.0
    else:
        print("Unknown traffic level. Assuming medium.")
        base_time *= 1.3
    weather = weather.lower()
    if weather == "rainy":
        base_time += 15
    elif weather == "stormy":
        base_time += 25
    elif weather == "clear":
        base_time += 0
    else:
        print("Unknown weather condition. Assuming clear.")

    base_time += 10


    final_eta = round(base_time)

    return final_eta
distance = float(input("Enter Distance (km): "))
traffic = input("Enter Traffic Level (Low / Medium / High): ")
weather = input("Enter Weather Condition (Clear / Rainy / Stormy): ")

eta = estimate_delivery_time(distance, traffic, weather)

print("\nEstimated Delivery Time:", eta, "minutes")
```

**Problem Statement 3:Movie Theatre Seat Occupancy**

```python
def analyze_booking(total_seats, booked_seats):
    booked_count = sum(booked_seats)

    occupancy = (booked_count / total_seats) * 10
    if occupancy == 100:
        status = "Housefull"
        suggestion = "Open an additional show immediately."
    elif occupancy >= 75:
        status = "Almost Full"
        suggestion = "Monitor closely. Consider opening another show."
    else:
        status = "Seats Available"
        suggestion = "No need for additional shows yet."
    print(f"Occupancy: {round(occupancy)}%")
    print(f"Show Status: {status}")
    print(f"Suggestion: {suggestion}")

total_seats = int(input("Enter Total Seats: "))

booked_number = int(input("Enter Number of Booked Seats: "))
booked_seats = [1] * booked_number

analyze_booking(total_seats, booked_seats)
```

Problem Statement 4:Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average_cpu = total / len(cpu_readings)
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {round(average_cpu)}%")
    print(f"Server Status: {status}")

readings_input = input("Enter CPU Readings (%) separated by commas: ")
cpu_readings = [int(value.strip()) for value in readings_input.split(",")]

classify_server_load(cpu_readings)
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def analyze_resource_usage(resource_usage):
    overused = []

    MAX_USAGE_LIMIT = 8

    for resource, hours in resource_usage.items():
        if hours > MAX_USAGE_LIMIT:
            overused.append(resource)

    # Display results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

analyze_resource_usage(resource_usage)
```
