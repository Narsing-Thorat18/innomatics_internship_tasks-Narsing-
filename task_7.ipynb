Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False

    def turn_on(self):
        self.is_on = True
        print(f"{self.name} is ON")

    def turn_off(self):
        self.is_on = False
        print(f"{self.name} is OFF")

    def status(self):
        state = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {state}")

light = SmartLight("Bedroom Light")
light.turn_on()
```

Problem 2: Employee ID Card System


```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

employee = Employee("Rahul", "EMP102", "IT")
employee.display_id_card()
```

Problem 3: Mobile Contact Record


```python
class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone_number)

contact_name = "Anita"
phone_number = "9876543210"

contact = Contact(contact_name, phone_number)
contact.display_contact()
```

Problem 4: Product Price Tag Generator


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹{}".format(self.price))

product_name = "Headphones"
price = 2499

product = Product(product_name, price)
product.print_price_tag()
```

Problem 5: Movie Rating Display System


```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

movie_name = "Inception"
rating = 4.8

movie = Movie(movie_name, rating)
movie.display_details()
```

Problem 6: Delivery Address Manager


```python
class Customer:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)

customer_name = "Suman"
address = "Hyderabad"

customer = Customer(customer_name, address)
customer.print_delivery_details()
```
