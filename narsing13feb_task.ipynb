 **Problem 1: Unique Words in a Sentence**

```python
# Taking input
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

**Problem 2: Highest Salary from Employee Data**

```python
#dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# employee with highest salary
highest_paid = max(employees, key=employees.get)
print(f"Highest Salary: {highest_paid} - {employees[highest_paid]}")
```

**Problem 3: Find Maximum and Minimum Values**

```python
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element
maximum = numbers[0]
minimum = numbers[0]
# Loop in list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

**Problem 4: Count Products Above a Price Threshold**

```python
# Input list
prices = [450, 1200, 899, 1500, 300]
# Initialize counter
count = 0
# Loop in list
for price in prices:
    if price > 1000:
        count += 1
# Output result
print("Products above 1000:", count)
```

**Problem 5: Calculate Attendance Percentage**

```python
# Input list
attendance = ["P", "P", "A", "P", "P"]

# Count total days
total_days = len(attendance)

# Count present days
present_count = 0
for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / total_days) * 100

# Output result
print("Attendance Percentage:", percentage)
```

**Problem 6: Remove Duplicate Phone Numbers**

```python
# Input list
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Output result
print("Unique phone numbers:", unique_numbers)
```

**Problem 7: Count Character Frequency**

```python
text = "pythonp"
char_count = {}
# Loop through each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
print(char_count)
```

**Problem 9: Check if a Key Exists in a Dictionary**

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Key check
key_name = "Ravi"
# Checking key existstance
if key_name in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
